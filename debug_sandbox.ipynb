```python
import numpy as np
import pytest
import cirq
import sympy
from cirq.contrib.custom_simulators.custom_state_simulator import CustomStateSimulator
from unittest import mock
from typing import List, Sequence, Tuple
from cirq import ops
import itertools
```

```python
#def test_state_copy():
sim = cirq.ClassicalStateSimulator()

q0, q1 = cirq.LineQubit.range(2)
simulator = cirq.ClassicalStateSimulator()
circuit = cirq.Circuit()
circuit.append(cirq.X(q0))
circuit.append(cirq.CNOT(q0, q1))
circuit.append(cirq.CNOT(q1, q0))
circuit.append(cirq.measure((q0, q1), key='key'))

state_Ms = []
for step in sim.simulate_moment_steps(circuit):
    state_Ms.append(step.state.M)
for x, y in itertools.combinations(state_Ms, 2):
    assert len(x) == len(y)
    for i in range(len(x)):
        assert not np.shares_memory(x[i], y[i])
```

```python

#q0, q1 = cirq.LineQubit.range(2)
simulator = cirq.ClassicalStateSimulator()
with mock.patch.object(simulator, '_core_iterator', wraps=simulator._core_iterator) as mock_sim:
    for b0 in [0, 1]:
        for b1 in [0, 1]:
            gate0 = cirq.X**b0
            gate1 = cirq.X**b1
            circuit = cirq.Circuit(
                gate0(q0), gate1(q1),cirq.measure(q0), cirq.measure(q1)
            )
            result = simulator.run(circuit, repetitions=3)
            np.testing.assert_equal(
                result.measurements, {'q(0)': [[b0]] * 3, 'q(1)': [[b1]] * 3}
            )
            assert result.repetitions == 3
    assert mock_sim.call_count == 8

```
